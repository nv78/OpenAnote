```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
bert_label_accuracies = pd.read_csv('TREC_Fine_label_ACC.csv')
ft_gpt_label_accuracies = pd.read_csv('FT_GPT3.5_fine_label_metrics_results.csv')
sefit_label_accuracies = pd.read_csv('setfit_label_acc.csv')
```

```python
sefit_label_accuracies
```

#  Accuracy, Precision, Recall Comparison Plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

# Plot Accuracy
plt.subplot(1, 3, 1)
plt.plot(bert_label_accuracies['Labels'], bert_label_accuracies['accuracy'], label='BERT', marker='o', linestyle='-', color='skyblue', linewidth=2)
plt.plot(ft_gpt_label_accuracies['Label Count'], ft_gpt_label_accuracies['Accuracy'], label='FT-GPT', marker='^', linestyle='-', color='coral', linewidth=2)
plt.plot(sefit_label_accuracies['Label'], sefit_label_accuracies['Accuracy'], label='Setfit', marker='^', linestyle='-.', color='lightcoral', linewidth=2)

plt.xlabel('Labels')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.legend()

# Plot Precision
plt.subplot(1, 3, 2)
plt.plot(bert_label_accuracies['Labels'], bert_label_accuracies['precision'], label='BERT', marker='o', linestyle='-', color='skyblue', linewidth=2)
plt.plot(ft_gpt_label_accuracies['Label Count'], ft_gpt_label_accuracies['Precision'], label='FT-GPT', marker='^', linestyle='-', color='coral', linewidth=2)
plt.plot(sefit_label_accuracies['Label'], sefit_label_accuracies['Precision'], label='Setfit', marker='^', linestyle='-.', color='lightcoral', linewidth=2)


plt.xlabel('Labels')
plt.ylabel('Precision')
plt.title('Precision Comparison')
plt.legend()

# Plot Recall
plt.subplot(1, 3, 3)
plt.plot(bert_label_accuracies['Labels'], bert_label_accuracies['recall'], label='BERT', marker='o', linestyle='-', color='skyblue', linewidth=2)
plt.plot(ft_gpt_label_accuracies['Label Count'], ft_gpt_label_accuracies['Recall'], label='FT-GPT', marker='^', linestyle='-', color='coral', linewidth=2)
plt.plot(sefit_label_accuracies['Label'], sefit_label_accuracies['Recall'], label='Setfit', marker='^', linestyle='-.', color='lightcoral', linewidth=2)

plt.xlabel('Labels')
plt.ylabel('Recall')
plt.title('Recall Comparison')
plt.legend()

plt.suptitle('Performance Comparison on Fine_Label TREC Dataset')

plt.tight_layout()
plt.show()

```
